```python
## Step_1: importing libraries ##

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
## Step_2: Creating a dataset ##
data = {
    'hours_studied':[1,2,3,4,5,6,7,8,9],
    'sleep_hours':[8,7,6,6,5,4,4,3,3],
    'marks_scored':[40,45,50,60,65,75,80,85,87]
}
df = pd.DataFrame(data)
df
```

```python
## Step_3: Defining features and targeted variables ##
x=df[['hours_studied','sleep_hours']]
y=df['marks_scored']
```

```python
## Step_4: Using train_test_split with 80% training and 20% testing data ##
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
## Step_5: Checking shapes of the training and testing sets ##
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
## Step_6: Making Predictions using Linear Regression ##
model=LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print("Predicted values:", y_pred) # More the training data,the predictions will be more accurate #
```

```python
## Step_7: Evaluating the model ##
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) # 
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
## Step_8: Visualizing the results ##
plt.scatter(x_test['hours_studied'], y_test, color = 'blue', label='Actual Data', marker='o')
plt.plot(x_test['hours_studied'], y_pred, color = 'red', label='Predicted Data')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Linear Regression: Actual vs Predicted')
plt.legend()
plt.show()
```

```python
## Step_9: Residual Analysis ##
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, color='blue', label='Residuals', marker='o')
plt.axhline(y=0, color='red',linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Analysis Graph')
plt.legend()
plt.show()
```

```python
## Step_10: Incpection of the model equation ##
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```
